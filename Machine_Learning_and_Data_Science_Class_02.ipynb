```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/classifier.csv")

df
```

```python
df.shape
```

```python
# Média
df[['Score1']].mean()
```

```python
# Mediana
df[['Score1']].median()
```

```python
# Desvio Padrão
df[['Score1']].std()
```

1) A geração de uma amostra não aleatória de 3000 linhas e a demonstração
de que não é adequada utilizá-la

```python
# não-aleatória
a1 = df.iloc[0:2999, :]
a1.shape
```

```python
a1[['Score1']].mean()
# 0.01%
```

```python
a1[['Score1']].median()
#  4.11%
```

```python
a1[['Score1']].std()
# 1% (acima)
```

```python
# Esta pode ser adadequada pois as estatísticas descritivas (média, mediana, desvio padrão) estão abaixo dos 5%
# Sendo assim, considerando a variável "Score1" do df, o resultado representa a população
```

2) A geração de uma amostra aleatória de 3000 linhas e a demonstração que é
adequado utilizá-la como amostra representativa

```python
# aleatória
a2 = df.sample(3000, random_state=4)
a2
```

```python
a2[['Score1']].mean()
# 3.1%
```

```python
a2[['Score1']].median()
# 2.2%
```

```python
a2[['Score1']].std()
# 0.95%
```

```python
# Esta também pode ser a forma adequada pois as estatísticas estão todas abaixo de 5%, porém, a amostra não-aleatória possui valores menores mostrando um nível de confiança maior.
# Sendo assim, considerando a variável "Score1" do df, o resultado também representa a população, mas, com um nível de confiança menor.
```
