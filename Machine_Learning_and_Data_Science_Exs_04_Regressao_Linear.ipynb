```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

Executar Regressões

```python
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/mtcars.csv")
df
```

Modelo mpg ~wt

```python
# Análise Exploratória

df.plot.scatter(x='mpg', y='wt')
```

```python
x = df[['mpg']]
y = df['wt']
# Análise de regressão
mod = LinearRegression()
mod.fit(x,y)

# R2
print(mod.score(x,y))
```

```python
# Visualização das estatísticas da regressão
mod_sm = sm.OLS(y,x).fit()
print(mod_sm.summary())
```

mpg ~ wt + cyl

```python
# Definindo as variáveis independentes e dependentes
x = df[['wt', 'cyl']]
y = df['mpg']

# Ajustando o modelo de regressão linear
mod = LinearRegression()
mod.fit(x, y)

# R2
print("R2 Score:", mod.score(x, y))

# Visualização das estatísticas da regressão
mod_sm = sm.OLS(y, sm.add_constant(x)).fit()
print(mod_sm.summary())

# Gerando valores para plotar a reta de regressão linear
x_fit = np.arange(1.5, 5.5, 0.05)
df_fit = pd.DataFrame({'wt': x_fit, 'cyl': np.mean(x['cyl'])})

y_fit = mod.predict(df_fit)

df_fit['mpg'] = y_fit

# Plotando o gráfico da regressão linear
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'], label='Dados Observados')
plt.plot(df_fit['wt'], df_fit['mpg'], color='red', label='Regressão Linear')
plt.xlabel('Weight (wt)')
plt.ylabel('MPG')
plt.title('Gráfico de Regressão Linear: MPG vs. Weight e Cylinders')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'], label='Dados Observados')
plt.xlabel('Weight (wt)')
plt.ylabel('MPG')
plt.title('Gráfico de Dispersão: MPG vs. Weight')
plt.legend()
plt.grid(True)
plt.show()

```
