```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

Executar Regress√µes

```python
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/mtcars.csv")
df
```

Modelo mpg ~wt

```python
# An√°lise Explorat√≥ria

df.plot.scatter(x='mpg', y='wt')
```

```python
x = df[['mpg']]
y = df['wt']
# An√°lise de regress√£o
mod = LinearRegression()
mod.fit(x,y)

# R2
print(mod.score(x,y))
```

```python
# Visualiza√ß√£o das estat√≠sticas da regress√£o
mod_sm = sm.OLS(y,x).fit()
print(mod_sm.summary())
```

mpg ~ wt + cyl

```python
# Definindo as vari√°veis independentes e dependentes
x = df[['wt', 'cyl']]
y = df['mpg']

# Ajustando o modelo de regress√£o linear
mod = LinearRegression()
mod.fit(x, y)

# R2
print("R2 Score:", mod.score(x, y))

# Visualiza√ß√£o das estat√≠sticas da regress√£o
mod_sm = sm.OLS(y, sm.add_constant(x)).fit()
print(mod_sm.summary())

# Gerando valores para plotar a reta de regress√£o linear
x_fit = np.arange(1.5, 5.5, 0.05)
df_fit = pd.DataFrame({'wt': x_fit, 'cyl': np.mean(x['cyl'])})

y_fit = mod.predict(df_fit)

df_fit['mpg'] = y_fit

# Plotando o gr√°fico da regress√£o linear
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'], label='Dados Observados')
plt.plot(df_fit['wt'], df_fit['mpg'], color='red', label='Regress√£o Linear')
plt.xlabel('Weight (wt)')
plt.ylabel('MPG')
plt.title('Gr√°fico de Regress√£o Linear: MPG vs. Weight e Cylinders')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o gr√°fico de dispers√£o
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'], label='Dados Observados')
plt.xlabel('Weight (wt)')
plt.ylabel('MPG')
plt.title('Gr√°fico de Dispers√£o: MPG vs. Weight')
plt.legend()
plt.grid(True)
plt.show()

```
