```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/base_transacional_supermercado_hash.csv")

df.head()
```

1) Agrupamentos temporais (MÃªs/Semana/Dia/12h-12h/8h-8h, etc)



```python
# converter a coluna data para modelo datetime
df['DATE'] = pd.to_datetime(df['DATE'])
df.head()
```

```python
# periodizar por mÃªs
df['mes'] = df['DATE'].dt.to_period('M')
df.head()
```

```python
# criar o df para mÃªs e agrupar por mÃªs
novo_df_mes = df.groupby('mes').apply(lambda x: x.reset_index(drop=True))
novo_df_mes
```

```python
# periodizar por semana
df['semana'] = df['DATE'].dt.week
df.head()
```

```python
# criar o data frame agrupado por semana
novo_df_semana = df.groupby('semana').apply(lambda x: x.reset_index(drop=True))
novo_df_semana
```

```python
# periodizar por dia
df['dia'] = df['DATE'].dt.date
df
```

```python
# criar o data frame agrupado por dia
novo_df_dia = df.groupby('dia').apply(lambda x: x.reset_index(drop=True))
novo_df_dia
```

2) Agrupamentos por Loja (Store_id)

```python
# criar o data frame de loja agrupado por Store ID
novo_df_loja = df.groupby('STORE_ID').apply(lambda x: x.reset_index(drop=True))
novo_df_loja.head()
```

3) Agrupamentos por (Ã¡rvore) de categorias (Category_1, 2, 3 e/ou 4)

```python
# criar o data frame de categoria 1 agrupado pela coluna CATEGORY_1
novo_df_categoria01 = df.groupby('CATEGORY_1').apply(lambda x: x.reset_index(drop=True))
novo_df_categoria01.head()
```

```python
# criar o data frame de categoria 02 agrupado pela coluna CATEGORY_2
novo_df_categoria02 = df.groupby('CATEGORY_2').apply(lambda x: x.reset_index(drop=True))
novo_df_categoria02.head()
```

```python
# criar o data frame de categoria 03 agrupado pela coluna CATEGORY_3
novo_df_categoria03 = df.groupby('CATEGORY_3').apply(lambda x: x.reset_index(drop=True))
novo_df_categoria03.head()
```

4) Experimentar diferentes funÃ§Ãµes de agregaÃ§Ã£o (soma, mÃ©dia, desvio
padrÃ£o, â¦)

```python
# criar o data frame de mes agrupado pelo mes demonstrando colunas com Soma, MÃ©dia e Desvio PadrÃ£o dos valores de Lucro Bruto
novo_df_mes.reset_index(inplace=True, drop=True)

novo_df_mes_agg = novo_df_mes.groupby('mes').agg({
    'GROSS_PROFIT_AMOUNT': ['sum', 'mean', 'std']})

novo_df_mes_agg
```

```python
# criar o data frame de semana agrupado pelo mes demonstrando colunas com Soma, MÃ©dia e Desvio PadrÃ£o dos valores de Lucro Bruto
novo_df_semana.reset_index(inplace=True, drop=True)

novo_df_semana_agg = novo_df_semana.groupby('semana').agg({
    'GROSS_PROFIT_AMOUNT': ['sum', 'mean', 'std']})

novo_df_semana_agg
```

```python
# criar o data frame de dia agrupado pelo mes demonstrando colunas com Soma, MÃ©dia e Desvio PadrÃ£o dos valores de Lucro Bruto
novo_df_dia.reset_index(inplace=True, drop=True)

novo_df_dia_agg = novo_df_dia.groupby('dia').agg({
    'GROSS_PROFIT_AMOUNT': ['sum', 'mean', 'std']})

novo_df_dia_agg
```

```python
# criar o data frame de dia agrupado por loja demonstrando colunas com Soma, MÃ©dia e Desvio PadrÃ£o dos valores de Lucro Bruto
novo_df_loja.reset_index(inplace=True, drop=True)

novo_df_loja_agg = novo_df_loja.groupby('STORE_ID').agg({
    'GROSS_PROFIT_AMOUNT': ['sum', 'mean', 'std']})

novo_df_loja_agg
```

```python
# criar o data frame de dia agrupado por categoria de produtos 01 demonstrando colunas com Soma, MÃ©dia e Desvio PadrÃ£o dos valores de Lucro Bruto

novo_df_categoria01.reset_index(inplace=True, drop=True)

novo_df_categoria01_agg = novo_df_categoria01.groupby('CATEGORY_1').agg({
    'GROSS_PROFIT_AMOUNT': ['sum', 'mean', 'std']})

novo_df_categoria01_agg
```
